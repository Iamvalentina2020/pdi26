umbralizaciones binaria y binaria inversa

```python
import cv2 
import matplotlib.pyplot as plt
imagen = cv2.imread('imagen.jpg')#lee en escalas de grises
#umbralizacion binaria
_, umbral_binario = cv2.threshold(imagen, 200, 255, cv2.THRESH_BINARY)
_, umbral_binaria_inv = cv2.threshold(imagen, 100, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')

plt.subplot(1, 3, 2)
plt.imshow(umbral_binario, cmap='gray')
plt.title('Umbral Binario')

plt.subplot(1, 3, 3)
plt.imshow(umbral_binaria_inv, cmap='gray')
plt.title('Umbral Binario Invertido')
plt.show()
```
