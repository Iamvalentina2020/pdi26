```python
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 3] 

img = cv2.imread('colores.jpg')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.subplot(141)
plt.title('Original')
plt.imshow(img_rgb)

plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(img_rgb[:,:,0], cmap='gray')

plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(img_rgb[:,:,1], cmap='gray')

plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(img_rgb[:,:,2], cmap='gray')

plt.show()
```

CON COLORES RELLENO

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 10]
img = cv2.imread('colores.jpg', cv2.IMREAD_COLOR)
imgRGB=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r=img[:,:,2]
g=img[:,:,1]
b=img[:,:,0]

#crear imagen por carnal

r_img=np.zeros_like(imgRGB)
r_img[:,:,0]=r

g_img=np.zeros_like(imgRGB)
g_img[:,:,1]=g

b_img=np.zeros_like(imgRGB)
b_img[:,:,2]=b


plt.figure(figsize=(15,5))


plt.subplot(141)
plt.title('Original')
plt.imshow(imgRGB)
plt.axis('off')

plt.subplot(142)
plt.title('Original')
plt.imshow(r_img)
plt.axis('off')

plt.subplot(143)
plt.title('Original')
plt.imshow(g_img)
plt.axis('off')

plt.subplot(144)
plt.title('Original')
plt.imshow(b_img)
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20, 8]

img_gray = cv2.imread('colores.jpg', cv2.IMREAD_GRAYSCALE)


plt.subplot(2, 5, 1)
plt.title('Falso Color: viridis')
plt.imshow(img_gray, cmap='viridis')
plt.axis('off')

plt.subplot(2, 5, 2)
plt.title('Falso Color: plasma')
plt.imshow(img_gray, cmap='plasma')
plt.axis('off')

plt.subplot(2, 5, 3)
plt.title('Falso Color: inferno')
plt.imshow(img_gray, cmap='inferno')
plt.axis('off')

plt.subplot(2, 5, 4)
plt.title('Falso Color: magma')
plt.imshow(img_gray, cmap='magma')
plt.axis('off')

plt.subplot(2, 5, 5)
plt.title('Falso Color: cividis')
plt.imshow(img_gray, cmap='cividis')
plt.axis('off')


plt.subplot(2, 5, 6)
plt.title('Falso Color: jet')
plt.imshow(img_gray, cmap='jet')
plt.axis('off')

plt.subplot(2, 5, 7)
plt.title('Falso Color: hot')
plt.imshow(img_gray, cmap='hot')
plt.axis('off')

plt.subplot(2, 5, 8)
plt.title('Falso Color: ocean')
plt.imshow(img_gray, cmap='ocean')
plt.axis('off')

plt.subplot(2, 5, 9)
plt.title('Falso Color: cool')
plt.imshow(img_gray, cmap='cool')
plt.axis('off')

plt.subplot(2, 5, 10)
plt.title('Falso Color: hsv')
plt.imshow(img_gray, cmap='hsv')
plt.axis('off')


plt.tight_layout() 
plt.show()
```
