# Imagen RGB con 4 Cuadrantes


```python
# 1. Importar librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Dimensiones y cálculo matemático de cuadrantes
N = 100
mitad = N // 2   # 100 // 2 = 50  → cada cuadrante es 50×50

print(f"Tamaño total  : {N}×{N}")
print(f"Mitad         : {mitad}")
print(f"Cada cuadrante: {mitad}×{mitad}")
```

```python
# 3. Canvas base con np.ones → lienzo blanco (valores 0–255)
imagen = np.ones((N, N, 3), dtype=np.uint8) * 255
print("Shape de la imagen:", imagen.shape)
```

```python
# 4. Rellenar cuadrantes con ciclos for
#    Cuadrante = (i // mitad) * 2 + (j // mitad)
#    0 → Superior-Izquierdo  | 1 → Superior-Derecho
#    2 → Inferior-Izquierdo  | 3 → Inferior-Derecho

colores = {
    0: [0,   255, 0  ],   # Verde   – Superior-Izquierdo
    1: [0,   0,   255],   # Azul    – Superior-Derecho
    2: [255, 0,   0  ],   # Rojo    – Inferior-Izquierdo
    3: [0,   255, 255],   # Celeste – Inferior-Derecho
}

for i in range(N):
    for j in range(N):
        cuadrante = (i // mitad) * 2 + (j // mitad)
        imagen[i, j] = colores[cuadrante]

print("Imagen rellenada correctamente")
```

```python
# 5. Visualizar y guardar
plt.figure(figsize=(4, 4))
plt.imshow(imagen)
plt.title("Imagen 100×100 – 4 Cuadrantes RGB")
plt.axis("off")
plt.tight_layout()
plt.savefig("imagen_cuadrantes.png", dpi=100)
plt.show()
```

# Imagen RGB con 9 Cuadrantes


```python
# 1. Importar librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Dimensiones para 9 cuadrantes (3x3)
N = 99  # Usamos 99 para que sea divisible exactamente entre 3
tercio = N // 3   # 99 // 3 = 33 → cada cuadrante es 33×33

print(f"Tamaño total   : {N}×{N}")
print(f"Tercio         : {tercio}")
print(f"Cada cuadrante : {tercio}×{tercio}")
```

```python
# 3. Canvas base (Lienzo blanco)
imagen = np.ones((N, N, 3), dtype=np.uint8) * 255
```

```python
# 4. Definir 9 colores para la grilla 3x3
colores = {
    0: [255, 0, 0],      # Rojo
    1: [0, 255, 0],      # Verde
    2: [0, 0, 255],      # Azul
    3: [255, 255, 0],    # Amarillo
    4: [255, 0, 255],    # Magenta
    5: [0, 255, 255],    # Cian
    6: [128, 0, 0],      # Marrón/Bordó
    7: [0, 128, 128],    # Verde Azulado
    8: [128, 128, 128],  # Gris
}
```

```python
# 5. Rellenar cuadrantes con ciclos for
# La fórmula cambia a: (i // tercio) * 3 + (j // tercio)
# Esto genera índices del 0 al 8 para una matriz de 3x3
for i in range(N):
    for j in range(N):
        # El multiplicador '3' define el ancho de la fila
        indice = (i // tercio) * 3 + (j // tercio)
        
        # Validación para evitar errores si N no es perfectamente divisible
        if indice in colores:
            imagen[i, j] = colores[indice]

print("Imagen de 9 cuadrantes rellenada correctamente")
```

```python
# 6. Visualizar y guardar
plt.figure(figsize=(5, 5))
plt.imshow(imagen)
plt.title(f"Imagen {N}x{N} - 9 Cuadrantes (3x3)")
plt.axis("off")
plt.tight_layout()
plt.savefig("imagen_9_cuadrantes.png", dpi=100)
plt.show()
```
