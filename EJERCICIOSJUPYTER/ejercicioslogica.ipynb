```python
#1 factorial
import math

try:
    n = int(input("1. Pon un número entero para factorial: "))
    print(f"Resultado: {math.factorial(n)}")
except ValueError:
    print("Error: Pon un número entero válido.")
```

```python
#2 primo
import math

try:
    n = int(input("2. Pon un número para verificar si es primo: "))
    if n < 2: 
        print("No es primo")
    else:
        es_primo = all(n % i != 0 for i in range(2, int(math.sqrt(n)) + 1))
        print(f"¿Es primo?: {'SÍ' if es_primo else 'NO'}")
except ValueError:
    print("Error: Entrada no válida.")
```

```python
#3 mayor y menor de 3 números
try:
    limite = int(input("3. ¿Cuántos números ingresarás? (mín 3): "))
    if limite < 3:
        print("Debes ingresar al menos 3 números.")
    else:
        nums = [float(input(f"Ingresa el número {i+1}: ")) for i in range(limite)]
        print(f"Mayor: {max(nums)} | Menor: {min(nums)}")
except ValueError:
    print("Error: Introduce solo números.")
```

```python
#4 palindromo (derecho y reves)
cadena = input("4. Introduce una palabra o frase: ").lower().replace(" ", "")
if cadena == "":
    print("Entrada vacía.")
else:
    resultado = "Es palíndromo" if cadena == cadena[::-1] else "No es palíndromo"
    print(f"Resultado: {resultado}")
```

```python
#5 el que mas se repite (frecuencia de 7 entradas)
from collections import Counter

print("5. Ingresa 7 números o palabras:")
try:
    nums = [input(f"Dato {i+1}: ") for i in range(7)]
    # most_common(1) devuelve una lista con un datito
    dato, cuenta = Counter(nums).most_common(1)[0]
    print(f"El valor que más se repite es '{dato}' ({cuenta} veces).")
except IndexError:
    print("Error al procesar los datos.")
```

```python
#6 anagrama (2 palabras tieen la misma letra)
p1 = input("6. Palabra 1: ").lower().replace(" ", "")
p2 = input("6. Palabra 2: ").lower().replace(" ", "")

if sorted(p1) == sorted(p2) and p1 != "":
    print("¡Es un anagrama!")
else:
    print("No es un anagrama.")
```

```python
#7 burbuja (menor a mayor comparando)
try:
    entrada = input("7. Números a ordenar (separados por espacio): ")
    nums = [float(x) for x in entrada.split()]
    
    n = len(nums)
    for i in range(n):
        for j in range(0, n - i - 1):
            if nums[j] > nums[j+1]:
                # Intercambio de valores (Sustitución simple en Python)
                nums[j], nums[j+1] = nums[j+1], nums[j]
    
    print(f"Lista ordenada: {nums}")
except ValueError:
    print("Error: Asegúrate de separar los números solo con espacios.")
```
